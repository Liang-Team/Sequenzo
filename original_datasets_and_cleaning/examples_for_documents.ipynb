```python
# Import necessary libraries
from sequenzo import * # Social sequence analysis
import pandas as pd # Import necesarry packages
```

```python
# Load the data that we would like to explore in this tutorial
# `df` is the short for `dataframe`, which is a common variable name for a dataset
df = load_dataset('country_co2_emissions')

# show
df
```

```python
# Ctrate SeqdataData
# Define the time-span variable
time = list(sequence_df_co2_emissions.columns)[1:]

states = ['Very Low', 'Low', 'Middle', 'High', 'Very High']

sequence_data = SequenceData(sequence_df_co2_emissions, time=time, states=states)

sequence_data
```

# 第一章：get_distance_matrix
这些是官方文档 get_distance_matrix 上的样例。

```python
# OMspell + TRATE
omspell = get_distance_matrix(sequence_data,
                         method="OMspell",
                         sm="TRATE",
                         indel="auto")
omspell
```

```python
# OM + CONSTANT
om = get_distance_matrix(sequence_data,
                         method="OM",
                         sm="CONSTANT",
                         indel="auto")
om
```

```python
# HAM + TRATE
ham = get_distance_matrix(sequence_data,
                          method="HAM",
                          sm="TRATE",
                          indel="auto")
ham
```

```python
# DHD + CONSTANT
dhd = get_distance_matrix(sequence_data,
                         method="DHD",
                         sm="CONSTANT",
                         indel="auto")
# Because DHD requires the 'TRATE' parameter to be used, the result is an error
dhd
```

# 第二章：get_substitution_matrix
这些是官方文档 get_substitution_matrix 上的样例。

```python
# TRATE + time_varying(True)
sm = get_substitution_cost_matrix(sequence_data,
                                  method="TRATE",
                                  cval=4,
                                  time_varying=True)
sm
```

```python
# CONSTANT + time_varying(False)
sm = get_substitution_cost_matrix(sequence_data,
                                  method="CONSTANT",
                                  cval=2,
                                  time_varying=False)
sm
```

# 第三章：clara
这是官方文档 clara 上的样例。

```python
# clara
result = clara(sequence_data,
               R=10,
               sample_size=3000,
               kvals=range(2,21),
               criteria=['distance', 'pbm'],
               parallel=True,
               stability=True)
result
```
