```python
#  timing, order/sequencing, duration 
#  LCPspell 
print(" LCPspell ...")
lcpspell_dist = get_distance_matrix(seqdata=sequence_data, method="LCPspell")
print("LCPspell ")

#  LCP 
print(" LCP ...")
lcp_dist = get_distance_matrix(seqdata=sequence_data, method="LCP")
print("LCP ")

import numpy as np
from collections import Counter

def analyze_sequence_features(seq_data, seq_id):
    """timing, order, duration"""
    #  numpy array  list
    if isinstance(seq_data.ids, np.ndarray):
        seq_idx = np.where(seq_data.ids == seq_id)[0][0]
    else:
        seq_idx = seq_data.ids.index(seq_id)
    seq_values = seq_data.values[seq_idx]
    
    # 
    valid_states = seq_values[seq_values >= 0]
    
    # 1. Order/Sequencing: 
    state_order = []
    prev_state = None
    for state in valid_states:
        if state != prev_state:
            state_order.append(state)
            prev_state = state
    
    # 2. Timing: 
    transitions = []
    prev_state = None
    for i, state in enumerate(valid_states):
        if state != prev_state and prev_state is not None:
            transitions.append(i)
        prev_state = state
    
    # 3. Duration: 
    durations = []
    current_state = valid_states[0]
    duration = 1
    for state in valid_states[1:]:
        if state == current_state:
            duration += 1
        else:
            durations.append((current_state, duration))
            current_state = state
            duration = 1
    durations.append((current_state, duration))
    
    return {
        'state_order': tuple(state_order),
        'transitions': transitions,
        'durations': durations,
        'num_transitions': len(transitions),
        'unique_states': set(valid_states)
    }

def compute_sequence_differences(feat1, feat2):
    """"""
    # Order difference: 
    order_diff = 0
    min_len = min(len(feat1['state_order']), len(feat2['state_order']))
    for i in range(min_len):
        if feat1['state_order'][i] != feat2['state_order'][i]:
            order_diff += 1
    order_diff += abs(len(feat1['state_order']) - len(feat2['state_order']))
    
    # Timing difference: 
    timing_diff = 0
    if feat1['transitions'] and feat2['transitions']:
        # 
        all_transitions = sorted(set(feat1['transitions'] + feat2['transitions']))
        if all_transitions:
            max_time = max(all_transitions)
            timing_diff = abs(feat1['num_transitions'] - feat2['num_transitions']) / max(1, max_time)
    
    # Duration difference: 
    duration_diff = 0
    state_durations1 = {state: dur for state, dur in feat1['durations']}
    state_durations2 = {state: dur for state, dur in feat2['durations']}
    all_states = set(state_durations1.keys()) | set(state_durations2.keys())
    for state in all_states:
        dur1 = state_durations1.get(state, 0)
        dur2 = state_durations2.get(state, 0)
        duration_diff += abs(dur1 - dur2)
    
    return {
        'order_diff': order_diff,
        'timing_diff': timing_diff,
        'duration_diff': duration_diff,
        'total_diff': order_diff + timing_diff * 10 + duration_diff  # 
    }

# 
print("\n...")
all_features = {}
for seq_id in sequence_data.ids:
    all_features[seq_id] = analyze_sequence_features(sequence_data, seq_id)

# 
print("\n...")
ref_candidates = []
for seq_id in sequence_data.ids:
    feat = all_features[seq_id]
    # 3-5
    if 3 <= feat['num_transitions'] <= 5:
        ref_candidates.append((seq_id, feat['num_transitions']))

# 
if ref_candidates:
    ref_id = ref_candidates[len(ref_candidates) // 2][0]  # 
else:
    ref_id = sequence_data.ids[0]  # 

print(f" ID = {ref_id}")
ref_features = all_features[ref_id]
# 
def get_state_label(seq_data, state_val):
    """"""
    try:
        state_val = int(state_val)
        
        # 1:  state_mapping
        if hasattr(seq_data, 'state_mapping') and seq_data.state_mapping is not None:
            mapping = seq_data.state_mapping
            if isinstance(mapping, dict):
                mapped_val = mapping.get(state_val)
                if mapped_val is not None:
                    # 
                    if isinstance(mapped_val, (int, np.integer)):
                        if 0 <= mapped_val < len(seq_data.labels):
                            return seq_data.labels[mapped_val]
                    else:
                        # 
                        return str(mapped_val)
        
        # 2:  state_val  labels 
        if 0 <= state_val < len(seq_data.labels):
            return seq_data.labels[state_val]
        
        # 3: 
        return f"State_{state_val}"
    except (ValueError, TypeError, IndexError) as e:
        return f"State_{state_val}(error)"

print(f"  : {[get_state_label(sequence_data, s) for s in ref_features['state_order']]}")
print(f"  : {ref_features['num_transitions']}")

# 
print("\n...")
differences = {}
for seq_id in sequence_data.ids:
    if seq_id != ref_id:
        diff = compute_sequence_differences(ref_features, all_features[seq_id])
        differences[seq_id] = diff

#  LCPspell 
dist_from_ref = lcpspell_dist.loc[ref_id].drop(ref_id, errors="ignore")
sorted_by_dist = dist_from_ref.sort_values()

# 
n = len(sorted_by_dist)
small_candidates = sorted_by_dist.head(max(10, n // 10)).index.tolist()
medium_candidates = sorted_by_dist.iloc[n//2 - 5:n//2 + 5].index.tolist()
large_candidates = sorted_by_dist.tail(max(10, n // 10)).index.tolist()

# 
def select_best_candidate(candidates, ref_feat, all_features, differences, mode='similar'):
    """
    
    mode='similar': - 
    mode='different': - 
    mode='medium':  - 
    """
    if not candidates:
        return None
    
    best_id = None
    
    if mode == 'similar':
        # 
        best_score = float('inf')
        for cand_id in candidates:
            if cand_id not in differences:
                continue
            diff = differences[cand_id]
            feat = all_features[cand_id]
            
            # 
            # 
            order_similarity = 0
            min_order_len = min(len(ref_feat['state_order']), len(feat['state_order']))
            for i in range(min_order_len):
                if ref_feat['state_order'][i] == feat['state_order'][i]:
                    order_similarity += 1
            
            # 
            # 
            score = -order_similarity * 100 + diff['total_diff']
            
            if score < best_score:
                best_score = score
                best_id = cand_id
    elif mode == 'different':
        # 
        best_score = -1
        for cand_id in candidates:
            if cand_id not in differences:
                continue
            diff = differences[cand_id]
            # 
            score = (diff['order_diff'] > 0) * 10 + \
                    (diff['timing_diff'] > 0.1) * 10 + \
                    (diff['duration_diff'] > 5) * 10 + \
                    diff['total_diff']
            if score > best_score:
                best_score = score
                best_id = cand_id
    else:  # mode == 'medium'
        # 
        all_scores = []
        for cand_id in candidates:
            if cand_id in differences:
                all_scores.append(differences[cand_id]['total_diff'])
        if all_scores:
            median_score = np.median(all_scores)
            best_score = float('inf')
            for cand_id in candidates:
                if cand_id not in differences:
                    continue
                score = abs(differences[cand_id]['total_diff'] - median_score)
                if score < best_score:
                    best_score = score
                    best_id = cand_id
    
    return best_id if best_id else (candidates[0] if candidates else None)

# 
small_id = select_best_candidate(small_candidates, ref_features, all_features, differences, mode='similar')
medium_id = select_best_candidate(medium_candidates, ref_features, all_features, differences, mode='medium')
large_id = select_best_candidate(large_candidates, ref_features, all_features, differences, mode='different')

# 
if not small_id:
    small_id = sorted_by_dist.index[0]
if not medium_id:
    medium_id = sorted_by_dist.index[len(sorted_by_dist) // 2]
if not large_id:
    large_id = sorted_by_dist.index[-1]

selected_ids = [ref_id, small_id, medium_id, large_id]

print("\n" + "="*60)
print(":")
print("="*60)

# 
ref_feat = all_features[ref_id]
print(f"\n (ID={ref_id}):")
print(f"  : {[get_state_label(sequence_data, s) for s in ref_feat['state_order']]}")
print(f"  : {ref_feat['num_transitions']}")
print(f"  : {[(get_state_label(sequence_data, s), d) for s, d in ref_feat['durations']]}")

# 
for i, (label, seq_id) in enumerate(zip(["", "", ""], [small_id, medium_id, large_id])):
    feat = all_features[seq_id]
    dist = lcpspell_dist.loc[ref_id, seq_id]
    diff = differences[seq_id]
    
    # 
    order_similarity = 0
    min_order_len = min(len(ref_feat['state_order']), len(feat['state_order']))
    for j in range(min_order_len):
        if ref_feat['state_order'][j] == feat['state_order'][j]:
            order_similarity += 1
    similarity_pct = (order_similarity / max(len(ref_feat['state_order']), len(feat['state_order']))) * 100 if min_order_len > 0 else 0
    
    print(f"\n{label} (ID={seq_id}):")
    print(f"  : {[get_state_label(sequence_data, s) for s in feat['state_order']]}")
    print(f"  : {feat['num_transitions']}")
    print(f"  : {[(get_state_label(sequence_data, s), d) for s, d in feat['durations']]}")
    print(f"  LCPspell: {dist:.4f}")
    print(f"  : {order_similarity}/{min_order_len} ({similarity_pct:.1f}%)")
    print(f"   - Order: {diff['order_diff']}, Timing: {diff['timing_diff']:.2f}, Duration: {diff['duration_diff']}")

print("\n" + "="*60)
print("''")
print("="*60)
```
