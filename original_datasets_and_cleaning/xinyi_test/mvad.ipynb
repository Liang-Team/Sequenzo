```python
def compare_distance_matrices(matrix1, matrix2, matrix1_name="Matrix 1", matrix2_name="Matrix 2", 
                             tolerance=1e-10, show_plots=True, max_sample_display=50):
    """
    Compare two distance matrices and provide detailed analysis.
    
    Parameters:
    - matrix1: First distance matrix (DataFrame or numpy array)
    - matrix2: Second distance matrix (DataFrame or numpy array)  
    - matrix1_name: Name for the first matrix (for display)
    - matrix2_name: Name for the second matrix (for display)
    - tolerance: Numerical tolerance for considering values as equal
    - show_plots: Whether to show visualization plots
    - max_sample_display: Maximum size for matrix visualization (to avoid memory issues)
    
    Returns: dict with comparison results
    """
    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt
    
    print("=" * 60)
    print("DISTANCE MATRIX COMPARISON ANALYSIS")
    print("=" * 60)
    print()
    
    # Convert to numpy arrays for easier comparison
    if hasattr(matrix1, 'values'):
        mat1 = matrix1.values
    else:
        mat1 = np.array(matrix1)
        
    if hasattr(matrix2, 'values'):
        mat2 = matrix2.values
    else:
        mat2 = np.array(matrix2)
    
    print(f"{matrix1_name} shape: {mat1.shape}")
    print(f"{matrix2_name} shape: {mat2.shape}")
    print()
    
    # Check if matrices have the same shape
    if mat1.shape != mat2.shape:
        print("❌ Matrices have different shapes!")
        return {
            'identical': False,
            'reason': 'shape_mismatch',
            'shape1': mat1.shape,
            'shape2': mat2.shape
        }
    
    # Check if matrices are identical
    are_identical = np.array_equal(mat1, mat2)
    print(f"Matrices are identical: {are_identical}")
    
    results = {
        'identical': are_identical,
        'shape1': mat1.shape,
        'shape2': mat2.shape
    }
    
    if not are_identical:
        # Calculate differences
        diff_matrix = mat1 - mat2
        
        # Statistics about differences
        print()
        print("=== DIFFERENCE STATISTICS ===")
        max_diff = np.max(np.abs(diff_matrix))
        mean_diff = np.mean(np.abs(diff_matrix))
        std_diff = np.std(diff_matrix)
        nonzero_count = np.count_nonzero(diff_matrix)
        nonzero_pct = nonzero_count / diff_matrix.size * 100
        
        print(f"Maximum difference: {max_diff:.6f}")
        print(f"Mean absolute difference: {mean_diff:.6f}")
        print(f"Standard deviation of differences: {std_diff:.6f}")
        print(f"Number of non-zero differences: {nonzero_count}")
        print(f"Percentage of non-zero differences: {nonzero_pct:.2f}%")
        
        # Check if differences are within tolerance
        are_close = np.allclose(mat1, mat2, atol=tolerance)
        print(f"Matrices are equal within tolerance {tolerance}: {are_close}")
        
        results.update({
            'max_difference': max_diff,
            'mean_difference': mean_diff,
            'std_difference': std_diff,
            'nonzero_differences': nonzero_count,
            'nonzero_percentage': nonzero_pct,
            'within_tolerance': are_close
        })
    
    print()
    print("=== MATRIX BASIC STATISTICS ===")
    print(f"{matrix1_name}:")
    print(f"  - Minimum value: {np.min(mat1):.6f}")
    print(f"  - Maximum value: {np.max(mat1):.6f}")
    print(f"  - Mean value: {np.mean(mat1):.6f}")
    print(f"  - Non-zero elements: {np.count_nonzero(mat1)}")
    
    print(f"{matrix2_name}:")
    print(f"  - Minimum value: {np.min(mat2):.6f}")
    print(f"  - Maximum value: {np.max(mat2):.6f}")
    print(f"  - Mean value: {np.mean(mat2):.6f}")
    print(f"  - Non-zero elements: {np.count_nonzero(mat2)}")
    
    results.update({
        'matrix1_stats': {
            'min': np.min(mat1),
            'max': np.max(mat1),
            'mean': np.mean(mat1),
            'nonzero_count': np.count_nonzero(mat1)
        },
        'matrix2_stats': {
            'min': np.min(mat2),
            'max': np.max(mat2),
            'mean': np.mean(mat2),
            'nonzero_count': np.count_nonzero(mat2)
        }
    })
    
    # Check if all values are zero (indicating potential issue)
    if np.all(mat1 == 0) and np.all(mat2 == 0):
        print()
        print("⚠️  WARNING: Both matrices contain only zeros!")
        print("   This might indicate an issue with distance calculation.")
        print("   Please check:")
        print("   1. Sequence data correctness")
        print("   2. Distance calculation parameters")
        print("   3. Whether all sequences are identical")
        results['warning'] = 'all_zeros'
    
    # Visualization
    if show_plots and not are_identical and (np.max(mat1) > 0 or np.max(mat2) > 0):
        print()
        print("=== VISUALIZATION ===")
        
        # Limit matrix size for visualization to avoid memory issues
        display_size = min(max_sample_display, mat1.shape[0])
        mat1_display = mat1[:display_size, :display_size]
        mat2_display = mat2[:display_size, :display_size]
        diff_display = mat1_display - mat2_display
        
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))
        
        # Plot 1: First matrix
        im1 = axes[0,0].imshow(mat1_display, cmap='viridis', aspect='auto')
        axes[0,0].set_title(f'{matrix1_name}')
        axes[0,0].set_xlabel('Sequence Index')
        axes[0,0].set_ylabel('Sequence Index')
        plt.colorbar(im1, ax=axes[0,0])
        
        # Plot 2: Second matrix
        im2 = axes[0,1].imshow(mat2_display, cmap='viridis', aspect='auto')
        axes[0,1].set_title(f'{matrix2_name}')
        axes[0,1].set_xlabel('Sequence Index')
        axes[0,1].set_ylabel('Sequence Index')
        plt.colorbar(im2, ax=axes[0,1])
        
        # Plot 3: Difference matrix
        im3 = axes[1,0].imshow(diff_display, cmap='RdBu_r', aspect='auto')
        axes[1,0].set_title(f'Difference ({matrix1_name} - {matrix2_name})')
        axes[1,0].set_xlabel('Sequence Index')
        axes[1,0].set_ylabel('Sequence Index')
        plt.colorbar(im3, ax=axes[1,0])
        
        # Plot 4: Histogram of differences
        diff_matrix = mat1 - mat2
        axes[1,1].hist(diff_matrix.flatten(), bins=50, alpha=0.7, edgecolor='black')
        axes[1,1].set_title('Distribution of Differences')
        axes[1,1].set_xlabel('Difference Value')
        axes[1,1].set_ylabel('Frequency')
        axes[1,1].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
        
        # Sample some specific values for detailed comparison
        print()
        print("=== SAMPLE VALUE COMPARISON ===")
        sample_indices = [(0, 1), (0, 10), (5, 15), (10, 20), (min(50, mat1.shape[0]-1), min(100, mat1.shape[1]-1))]
        print(f"Position (i,j)     | {matrix1_name:>12} | {matrix2_name:>12} | Difference")
        print("-" * 60)
        for i, j in sample_indices:
            if i < mat1.shape[0] and j < mat1.shape[1]:
                val1 = mat1[i, j]
                val2 = mat2[i, j]
                diff = val1 - val2
                print(f"({i:2d},{j:2d})            | {val1:8.4f}     | {val2:8.4f}     | {diff:8.4f}")
        
    elif are_identical:
        print()
        print("Matrices are identical - no visualization needed")
        
        # Even if identical, show a sample of the matrix structure
        if show_plots and np.max(mat1) > 0:
            display_size = min(max_sample_display, mat1.shape[0])
            plt.figure(figsize=(8, 6))
            plt.imshow(mat1[:display_size, :display_size], cmap='viridis', aspect='auto')
            plt.title(f'Distance Matrix Structure (First {display_size}x{display_size} subset)')
            plt.xlabel('Sequence Index')
            plt.ylabel('Sequence Index')
            plt.colorbar()
            plt.show()
    
    return results
```

```python
def compare_summary_tables(weighted_table, unweighted_table, tolerance=1e-6):
    """
    Compare weighted and unweighted summary tables to identify differences.
    
    Parameters:
    - weighted_table: DataFrame with weights
    - unweighted_table: DataFrame without weights  
    - tolerance: numerical tolerance for considering values equal
    
    Returns: dict with comparison results
    """
    import pandas as pd
    import numpy as np
    
    print("=" * 50)
    print("SUMMARY TABLE COMPARISON")
    print("=" * 50)
    
    # Basic structure check
    print("\n1. Structure Check:")
    print(f"Weighted shape:   {weighted_table.shape}")
    print(f"Unweighted shape: {unweighted_table.shape}")
    
    shape_match = weighted_table.shape == unweighted_table.shape
    columns_match = list(weighted_table.columns) == list(unweighted_table.columns)
    
    if not shape_match:
        print("❌ Different shapes!")
        return {'identical': False, 'reason': 'shape_mismatch'}
    
    if not columns_match:
        print("❌ Different columns!")
        weighted_only = set(weighted_table.columns) - set(unweighted_table.columns)
        unweighted_only = set(unweighted_table.columns) - set(weighted_table.columns)
        if weighted_only:
            print(f"  Only in weighted: {list(weighted_only)}")
        if unweighted_only:
            print(f"  Only in unweighted: {list(unweighted_only)}")
        return {'identical': False, 'reason': 'column_mismatch'}
    
    print("✅ Same structure")
    
    # Value comparison
    print("\n2. Value Comparison:")
    differences = []
    identical = True
    
    for col in weighted_table.columns:
        try:
            # Try numerical comparison first
            w_vals = pd.to_numeric(weighted_table[col], errors='coerce')
            u_vals = pd.to_numeric(unweighted_table[col], errors='coerce')
            
            if not w_vals.isna().all():  # numerical column
                diff = np.abs(w_vals - u_vals)
                significant_diffs = diff > tolerance
                
                if significant_diffs.any():
                    identical = False
                    max_diff = diff.max()
                    mean_diff = (w_vals - u_vals).mean()
                    pct_diff = (significant_diffs.sum() / len(diff)) * 100
                    
                    differences.append({
                        'column': col,
                        'max_diff': max_diff,
                        'mean_diff': mean_diff,
                        'pct_different': pct_diff
                    })
                    
                    print(f"{col}: {significant_diffs.sum()}/{len(diff)} different ({pct_diff:.1f}%)")
                    print(f"  Max difference: {max_diff:.6f}")
                    print(f"  Mean difference: {mean_diff:.6f}")
                else:
                    print(f"{col}: ✅ identical")
            else:  # categorical column
                matches = weighted_table[col] == unweighted_table[col]
                if not matches.all():
                    identical = False
                    pct_diff = ((~matches).sum() / len(matches)) * 100
                    print(f"{col}: {(~matches).sum()}/{len(matches)} different ({pct_diff:.1f}%)")
                else:
                    print(f"{col}: ✅ identical")
                    
        except Exception as e:
            print(f"{col}: Error - {e}")
    
    # Summary
    print(f"\n3. Summary:")
    if identical:
        print("✅ Tables are IDENTICAL")
    else:
        print("❌ Tables are DIFFERENT")
        print(f"   Columns with differences: {len(differences)}")
        
        if differences:
            biggest_diff = max(differences, key=lambda x: x['max_diff'])
            print(f"   Biggest difference in '{biggest_diff['column']}': {biggest_diff['max_diff']:.6f}")
    
    return {
        'identical': identical,
        'differences': differences,
        'summary': 'identical' if identical else f'{len(differences)} columns differ'
    }
```
