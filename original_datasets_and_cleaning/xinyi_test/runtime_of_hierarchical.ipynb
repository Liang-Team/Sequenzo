# Time

```python
import time

import numpy as np
import pandas as pd
from sequenzo import *
from sequenzo.clustering import Cluster
```

## Broad_df_5_years

```python
# get dissimilarity distance matrix
df = pd.read_csv("D:/college/research/QiQi/sequenzo/files/sampled_data_sets/broad_data/sampled_5000_data.csv")   # 每次修改中间的 ‘5000’ 即可

_time = list(df.columns)[4:]

states = ['Non-computing', 'Non-technical computing', 'Technical computing']

df = df[['worker_id', 'C1', 'C2', 'C3', 'C4', 'C5']]

sequence_data = SequenceData(df, time_type="age", time=_time, id_col="worker_id", states=states)

diss = get_distance_matrix(sequence_data, method="OMspell", sm="TRATE", indel="auto")

diss
```

```python
# hierarchical clustering
start_time = time.time()
cluster = Cluster(diss, sequence_data.ids, clustering_method='ward')
print("step1 success.")
cluster.plot_dendrogram(xlabel="Countries", ylabel="Distance", save_as='dendrogram')
print("step2 success.")
end_time = time.time()
print(f"{end_time - start_time} seconds")
```

## Detailed_df_10_years

```python
# get dissimilarity distance matrix
df = pd.read_csv('D:/college/research/QiQi/sequenzo/files/sampled_datasets/detailded_data/sampled_1000_data.csv')   # # 每次修改中间的 ‘1000’ 即可

_time = list(df.columns)[4:]
states = ['data', 'data & intensive math', 'hardware', 'research', 'software', 'software & hardware', 'support & test']
df = df[['worker_id', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10']]

data = SequenceData(df, time=_time, time_type="age", id_col="worker_id", states=states)
diss = get_distance_matrix(seqdata=data, method="OMspell", sm="TRATE", indel="auto")
```

```python
# hierarchical clustering
start_time = time.time()
cluster = Cluster(diss, data.ids, clustering_method='ward')
cluster.plot_dendrogram(xlabel="Countries", ylabel="Distance", save_as='dendrogram')
end_time = time.time()
print(f"{end_time - start_time} seconds")
```

# Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 绘制折线图
def getPlot(data, save_path):
    df = pd.DataFrame(data)

    # Plot setup
    plt.figure(figsize=(10, 6))
    sns.set(style="whitegrid")

    # Plotting both R and Python data
    plt.plot(df['datasize'], df['R'], marker='o', label='R TraMineR', color='#5698C3', linewidth=2, markersize=8)
    plt.plot(df['datasize'], df['python'], marker='s', label='Python Sequenzo', color='#FCA104', linewidth=2,
             markersize=8)

    # Title and labels with more padding
    plt.xlabel('Data Size (in 500s)', fontsize=14, labelpad=35)
    plt.ylabel('Execution Time (seconds)', fontsize=14, labelpad=35)

    # Adding a legend with the specified caption
    plt.legend(title="Legend", title_fontsize='13', loc='upper left')

    # Adding a grid for better readability
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)

    # Adjusting layout to ensure there's no clipping of labels or titles
    plt.tight_layout()

    plt.savefig(save_path, dpi=300, bbox_inches='tight')

    plt.show()
```
