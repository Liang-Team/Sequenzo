```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# 绘制折线图
def getPlot(data, save_path, x_axis, x_label, baseline=None, note=None, y_label="Execution Time (seconds)", colors=['#ffadbb', '#acbfeb']):
    data = pd.DataFrame(data, index=x_axis)

    # Plot setup
    plt.figure(figsize=(10, 6))
    sns.set(style="whitegrid")

    # Plotting both R and Python data
    for (label, series), color in zip(data.items(), colors):
        # 判断是不是基准线
        if baseline is not None and any(keyword in label for keyword in baseline):
                plt.plot(data.index, series, linestyle='--', color='black', linewidth=1.2, label=label)
        else:
            # 普通折线
            plt.plot(data.index, series, marker='o', label=label, color=color, linewidth=2, markersize=8)

    # Title and labels with more padding
    plt.xlabel(x_label, fontsize=14, labelpad=35)
    plt.ylabel(y_label, fontsize=14, labelpad=35)

    plt.xticks(x_axis)

    # Adding a legend with the specified caption
    plt.legend(title="Legend", title_fontsize='13', loc='upper left')

    # Adding a grid for better readability
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)

    # Adjusting layout to ensure there's no clipping of labels or titles
    plt.tight_layout()

    # Add note under the plot
    if note is not None:
        plt.figtext(0.54, -0.05, note, wrap=True, horizontalalignment='center', fontsize=10, style='italic')

    plt.savefig(save_path, dpi=300, bbox_inches='tight')

    # plt.show()
```

```python
def getPlot_unique(data_dict, save_path, x_axis, colors=['#ffadbb', '#acbfeb', '#e6ccff']):
    df = pd.DataFrame(data_dict, index=x_axis)

    plt.figure(figsize=(10, 6))
    sns.set(style="whitegrid")

    # colors = ["#FF6FB5", "#FFA559", "#FFE45E", "#9DFF8B", "#7EC8E3", "#CBA1FF", "#7FFFD4"]

    for (label, series), color in zip(df.items(), colors):
        plt.plot(df.index, series, marker='o', label=label,
                 color=color, linewidth=2, markersize=8)

    plt.xlabel('Data Size (10,000)', fontsize=14, labelpad=20)
    plt.ylabel('Execution Time (seconds)', fontsize=14, labelpad=20)

    plt.xticks(x_axis)

    plt.legend(title="Test", title_fontsize=13, loc='upper left')

    plt.grid(True, which='both', linestyle='--', linewidth=0.5)

    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
```
