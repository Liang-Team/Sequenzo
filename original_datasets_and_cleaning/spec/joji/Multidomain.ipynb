```python


# Get Supressed_Viral_Load_Prevalence columns
svl_prevalence_cols = [col for col in df.columns if col.startswith('Supressed_Viral_Load_Prevalence')]

print(svl_prevalence_cols)

# Get Incident_Rate columns
incident_rate_cols = [col for col in df.columns if col.startswith('Incident_Rate')]
print(incident_rate_cols)
```

```python
# Create new DataFrames
df_svl_prevalence = df[['Country'] + svl_prevalence_cols]
df_incident_rate = df[['Country'] + incident_rate_cols]
df_incident_rate
```

```python
df_svl_prevalence = pd.read_csv("svl_prevalence.csv")
df_incident_rate= pd.read_csv("incident_rate.csv")
df_incident_rate
```

```python
df_svl_prevalence.columns = df_svl_prevalence.columns.str.extract(r'(\d{4})')[0].
df_incident_rate.columns = df_incident_rate.columns.str.extract(r'(\d{4})')[0]
```

```python
df_svl_prevalence.rename (columns={'Unnamed: 0': 'Country'}, inplace=True)
df_incident_rate.rename (columns={'Unnamed: 0': 'Country'}, inplace=True)
```
