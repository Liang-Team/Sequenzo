## Dissimilarity Measures

```python
# You can also replace "OMspell" with "OM/DHD/HAM" and "TRATE" with "CONSTANT"
om = get_distance_matrix(seqdata=sequence_data,
                         method='OM',
                         sm="TRATE",
                         indel="auto")
om
```

## Cluster Analysis

```python
cluster = Cluster(om, sequence_data.ids, clustering_method='ward_d2')
cluster.plot_dendrogram(xlabel="Countries", ylabel="Distance")
```

```python
# Create a ClusterQuality object to evaluate clustering quality
cluster_quality = ClusterQuality(cluster)
cluster_quality.compute_cluster_quality_scores()
cluster_quality.plot_cqi_scores(norm='zscore', save_as='combined_scores')
summary_table = cluster_quality.get_cqi_table()
print(summary_table)
```

```python
cluster_results = ClusterResults(cluster)
membership_table = cluster_results.get_cluster_memberships(num_clusters=5)
print(membership_table)
distribution = cluster_results.get_cluster_distribution(num_clusters=5)
print(distribution)
cluster_results.plot_cluster_distribution(num_clusters=6, save_as="distribution.png", title=None)
```

## Regression

Now that we have identified clusters of emission trajectories, we may ask: what sequence-level characteristics distinguish the clusters?

We use Sequenzo's built-in sequence characteristics (number of transitions, within-sequence entropy, turbulence, and complexity index) and fit a multinomial logistic regression to predict cluster membership from these features.

```python
# Step 1: Compute sequence characteristics
from sequenzo.sequence_characteristics import (
    get_number_of_transitions,
    get_within_sequence_entropy,
    get_turbulence,
    get_complexity_index,
)

trans      = get_number_of_transitions(sequence_data, norm=True)
entropy    = get_within_sequence_entropy(sequence_data, norm=True)
turb       = get_turbulence(sequence_data, norm=True, type=1, id_as_column=True)
complexity = get_complexity_index(sequence_data)

print("complexity columns:", complexity.columns.tolist())

df_reg = pd.DataFrame({
    'Transitions': trans['Transitions'].values,
    'Entropy': entropy['Entropy'].values,
    'Turbulence': turb['Turbulence'].values,
    'Complexity': complexity['Complexity Index'].values,
    'Cluster': membership_table['Cluster'].values,
})
df_reg.head()
```

```python
# Step 2: Descriptive statistics by cluster
feature_cols = ['Transitions', 'Entropy', 'Turbulence', 'Complexity']
print("Mean sequence characteristics by cluster:")
print(df_reg.groupby('Cluster')[feature_cols].mean().round(3))
```

```python
# Step 3: Multinomial logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

X = StandardScaler().fit_transform(df_reg[feature_cols])
y = df_reg['Cluster'].values

model = LogisticRegression(solver='lbfgs', max_iter=1000, random_state=42)
model.fit(X, y)

print(f"Training accuracy: {model.score(X, y):.1%}")
print("\nCoefficients (standardized):")
print(pd.DataFrame(model.coef_, index=[f"Cluster {i}" for i in model.classes_],
                    columns=feature_cols).round(3))
```

## Conclusions

In this tutorial, we performed a complete sequence analysis workflow on country-level CO₂ emissions data (194 countries, 1800-2022), where each country's annual emissions are classified into five levels from Very Low to Very High based on global quintile thresholds.

1. **Data exploration**: We examined individual country trajectories (e.g., Andorra) and the overall state distribution. Most countries started at Very Low emissions and diverged over time. By 2022, 125 out of 194 countries are classified as Very High.

2. **Visualization**: Index plots showed the full heterogeneity of emission trajectories across countries, while state distribution plots revealed the global shift from low to high emissions over two centuries.

3. **Dissimilarity and clustering**: Using Optimal Matching with transition-rate substitution costs, we computed pairwise dissimilarity between all country trajectories and applied Ward's hierarchical clustering to identify 5 trajectory types.

4. **Regression**: We computed four sequence characteristics (transitions, entropy, turbulence, complexity) using Sequenzo's built-in functions, then used multinomial logistic regression to understand what distinguishes the clusters. This links the sequence typology back to measurable properties of each trajectory.

This workflow illustrates how Sequenzo supports a complete analysis pipeline for social sequence analysis, from raw sequences through visualization, clustering, and regression.
